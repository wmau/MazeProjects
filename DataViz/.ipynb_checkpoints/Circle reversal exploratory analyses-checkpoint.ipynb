# How many ensembles are in young vs aged mice?

```python
# Count raw number of ensembles per mouse. Don't normalize by number of recorded neurons per session. 
n_ensembles = RR.count_ensembles(normalize=False)
```

```python
# No difference in number of ensembles in young vs aged mice. 
for session in RR.meta['session_types']:
    p = np.round(
        mannwhitneyu(
            n_ensembles[session]['young'], 
            n_ensembles[session]['aged']
        ).pvalue, 
        2)
    msg = f'{session}: p = {p}'
    if p < 0.05:
        msg += '*' 
    print(msg)
```

```python
# Count raw number of ensembles per mouse. Normalize by number of recorded neurons per session. 
n_ensembles_norm = RR.count_ensembles(normalize=True)
```

```python
# No difference in number of ensembles in young vs aged mice. 
for session in RR.meta['session_types']:
    p = np.round(
        mannwhitneyu(
            n_ensembles_norm[session]['young'], 
            n_ensembles_norm[session]['aged']
        ).pvalue,
        2)
    msg = f'{session}: p = {p}'
    if p < 0.05:
        msg += '*' 
    print(msg)
```

# What is the composition of ensembles in young vs aged mice?

```python
# Example ensemble. Red neurons on the right are those with a weight > 2 s.d. above the mean. Only those neurons are plotted
# in the raster on the left. 
fig, ax = RR.data['Fornax']['Goals4'].plot_assembly(0)
```

```python
# Each mouse has ~60 ensembles, so for each of those ensembles, count how many neurons pass the threshold to be "members". 
# This gives you a distribution of ensemble sizes per mouse. Take the median of that distribution to get the distributions below.
# Each boxplot contains one data point from each mouse. 
ensemble_makeup = RR.plot_ensemble_sizes(data_type='ensemble_size')
```

```python
# Ensemble size mostly the same across ages. 
for session in RR.meta['session_types']:
    p = np.round(
        mannwhitneyu(
            ensemble_makeup['ensemble_size'][session]['young'], 
            ensemble_makeup['ensemble_size'][session]['aged']
        ).pvalue, 
        2)
    msg = f'{session} p = {p}'
    if p < 0.05:
        msg += '*' 
    print(msg)
```

```python
# For each ensemble, find its members. Counting all these members, what proportion of the total population do these cells 
# make up? They're called "unique" members because cells that belong to multiple ensembles only get counted once. 
ensemble_makeup = RR.plot_ensemble_sizes(data_type='unique_members')
```

```python
# Trending significance. Might indicate that in aged mice, there are slightly fewer . Under our definition of "ensemble" this
# means fewer neurons are co-active in aged mice, but the effect size is small. Will pursue more straightforward way of asking 
# this. Also, I should next look whether aged neurons are more "promiscuous" (i.e., belong to multiple ensembles) or 
# whether aged mice simply have fewer neurons in their ensembles. The analysis above is not significant but is underpowered. 
for session in RR.meta['session_types']:
    p = np.round(
        mannwhitneyu(
            ensemble_makeup['unique_members'][session]['young'], 
            ensemble_makeup['unique_members'][session]['aged']
        ).pvalue, 
        2)
    msg = f'{session}: p = {p}'
    if p < 0.05:
        msg += '*' 
    print(msg)
```

# Do ensembles have spatial selectivity?

```python
# Example ensemble with a clear spatial field. Green line corresponds to rewarded port. 
ax = RR.data['Fornax']['Goals3'].spiralplot_assembly(31, threshold=0)
```

```python
# Take the median spatial information across ensembles for each mouse. The spatial information is z-scored relative to 
# shuffled controls. This is better than raw spatial information. See https://www.sciencedirect.com/science/article/abs/pii/S0306452218301088?via%3Dihub.

# To make the shuffled distributions:
# 1. Shuffle ensemble activity in time by a random amount. 
# 2. Compute spatial information of this shuffled ensemble. 
# 3. Repeat #1-2 500 times to get 500 randomized spatial information values. 
# 4. Z-score the real spatial information of the unshuffled ensemble relative to that distribution. 

SI = RR.boxplot_all_assembly_SI()
```

```python
# First three sessions are trending significance. Tentatively, this can be interpreted as aged mice take longer to catch up to 
# young mice in terms of spatial selectivity. 
for session in RR.meta['session_types']:
    p = np.round(
        mannwhitneyu(
            SI[session]['young'], 
            SI[session]['aged']
        ).pvalue, 
        2)
    msg = f'{session}: p = {p}'
    if p < 0.05:
        msg += '*' 
    print(msg) 
```

# How similar are ensembles across sessions?

```python
# By using "ensemble registration" we can tentatively match ensembles across days. 
# To do this: 
# 1. Pick two sessions to register across. 
# 2. In session 1, get an ensemble's weight vector. 
# 3. Drop all neurons that are not common across the two sessions. 
# 4. Reorder the weight vectors of session 2 so that the neurons correspond to the same registered neurons as session 1. 
# 5. For each ensemble in session 2, compute its cosine similarity with the ensemble in session 1. 
# 6. Find the highest cosine similarity. If its z-score (relative to all the cosine similarities) is > 2.58, it's considered a match. 
# 7. Repeat #1-6 for all ensembles in session 1. 

# One example highly matched ensemble. 
RR.plot_matched_ensembles('Fornax', ('Goals3', 'Goals4'), subset=[31])
```

```python
# What percentage of ensembles register across sessions? Is there a difference across ages? 
# Most ensembles register successfully ("match").
p_matches = RR.percent_matched_ensembles()
```

```python
# Do ensembles match more across Goals3-Goals4 compared to Goals4-Reversal? 
# Compare match %s within mice with those two session pairs. 
# No difference in match % regardless of whether or not there was a reversal. 
for age in ['young', 'aged']:
    p = np.round(
        wilcoxon(
            p_matches[age][('Goals3', 'Goals4')], 
            p_matches[age][('Goals4', 'Reversal')]
        ).pvalue, 
        2)
    msg = f'{age} Goals3-Goals4 match % vs Goals4-Reversal match %: p = {p}'
    if p < 0.05:
        msg += '*' 
    print(msg)
```

```python
# Is there a difference in age for match percentages?
# No difference in match % across ages.
for session_pair in [('Goals3', 'Goals4'), ('Goals4', 'Reversal')]:
    p = np.round(
        mannwhitneyu(
            p_matches['young'][session_pair], 
            p_matches['aged'][session_pair]
        ).pvalue, 
        2)
    msg = f'match % for {session_pair}, young vs aged: p = {p}'
    if p < 0.05:
        msg += '*' 
    print(msg)
```

```python
# Of the ensembles that matched, how similar were their weight vectors? Compare Goals3-Goals4 to Goals4-Reversal. Perhaps
# Goals3-Goals4 would have more similar ensembles? 
session_pair1 = ('Goals3', 'Goals4') # left boxplots above each mouse
session_pair2 = ('Goals4', 'Reversal') #right boxplots above each mouse

registration_similarities = RR.plot_pattern_cosine_similarity_comparisons(
    session_pair1=session_pair1, 
    session_pair2=session_pair2
)
```

```python
# Ensembles are just as similar in Goals3-Goals4 as they are in Goals4-Reversal. 
for age in ['young', 'aged']:
    for sims1, sims2, mouse in zip(registration_similarities[session_pair1][age], 
                                   registration_similarities[session_pair2][age],
                                   RR.meta['grouped_mice'][age]):
        p = np.round(
            mannwhitneyu(sims1, sims2).pvalue, 
            2)
        msg = f'{age} {mouse}: p = {p}'
        if p < 0.05:
            msg += '*'         
        print(msg)
```

# What is the spatial correlation of ensembles across days?

```python
# Get spatial fields of the ensembles across days. They match pretty well. 
# Red lines correspond to reward location. 
fields1 = RR.snakeplot_matched_ensembles('Fornax', ('Goals3', 'Goals4'))
fields2 = RR.snakeplot_matched_ensembles('Fornax', ('Goals4', 'Reversal'))
```

```python
# Correlate the ensemble fields across days. Same as above, Goals3-Goals4 correlations are the left boxplots and Goals4-Reversal
# correlations are the right boxplots. We might expect Goals3-Goals4 correlations to be higher since there could be ensemble
# remapping during Reversal. 
ensemble_field_rhos = RR.plot_ensemble_field_correlation_comparisons(session_pair1, session_pair2)
```

```python
# Only one young mouse has a drop in correlation from Goals3-Goals4 to Goals4-Reversal. 
# One aged mouse has an increase in correlation. 
for age in ['young', 'aged']:
    for sims1, sims2, mouse in zip(ensemble_field_rhos[session_pair1][age], 
                                   ensemble_field_rhos[session_pair2][age],
                                   RR.meta['grouped_mice'][age]):
        p = np.round(
            mannwhitneyu(sims1, sims2).pvalue, 
            2)
        msg = f'{age} {mouse}: p = {p}'
        if p < 0.05:
            msg += '*' 
        print(msg)
```
